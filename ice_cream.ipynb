```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Ice_Cream_Ratings.csv")
df = df.set_index('Date')
```

```python
df
```

```python
# default gives a line plot
df.plot()
```

```python
# splits the column values into their own visualisation
df.plot(kind = 'line', subplots = True)
```

```python
# STYLING THE CHART - FIRST PRINT THIS COMMAND TO SEE ALL THE STYLES POSSIBLE
print(plt.style.available)
#plt.style.use('seaborn-v0_8')
plt.style.use('seaborn-v0_8-deep')

```

```python
# X & Y label titles with a title
df.plot(kind = 'line', title = "Ice Cream Ratings", xlabel= 'Daily Ratings', ylabel = 'Scores')
```

```python
# creating a bar chart
## True makes them stacked into one
## False makes them separate side by side
df.plot(kind = 'bar', title = "Ice Cream Ratings", stacked = True)
```

```python
# Create BARCHART with just one of the columns / flavour types
df['Flavor Rating'].plot(kind = 'bar', title = "Ice Cream Ratings", stacked = True)
```

```python
# Create a horizonal BARCHART, stacked with heading
df.plot.barh(title = "Ice Cream Ratings", stacked = True)
```

```python
# Create a SCATTER PLOT, we need to define the X & Y values
# PARAMETERS
# s = int // changes size of dots

df.plot.scatter(x='Texture Rating', y = 'Overall Rating', s=100)
```

```python
# Create a HISTOGRAM 
# PARAMETERS
# bins = int // to show number of bins

df.plot.hist(bins = 20)
```

```python
# Creating a BOXPLOT
# whiskers, showing max and min outliers
# Bottom of the box = 25 percentile
# Green horiz line within the box = Mean (50%)
# Top of the box = 75 percentile

df.boxplot()
```

```python
# PLOT AREA with modified figsize, which we can use on all charts
df.plot.area(figsize = (10,5))
```

```python
# PIECHART
# we need to pass y value
df.plot.pie(y='Flavor Rating', figsize = (10,5))
```
